```python
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox

def dectect_and_draw(filename,model="yolov3-tiny",confidence=0.5):
  image_file = f"{filename}"
  img = cv2.imread(image_file)
  bbox,label,conf = cv.detect_common_objects(img,confidence=confidence,model=model)
  print(f"image processed is {filename}\n")
  for l,c in zip(label,conf):
    print(f"deted object is {l} with confidence of {c}")
  out_image = draw_bbox(img,bbox,label,conf)
  cv2.imwrite(f"{filename}",out_image)
  display(Image(filename=f"{filename}"))
```

```python
for file_name in image_files:
  dectect_and_draw(file_name)
```
